#### Clean Data

```python
#create separatae date and time columns

_date = []
_time = []

for values in data['endTime']:
    _date.append(re.search("\d{4}-\d{2}-\d{2}", values).group())

for values in data['endTime']:
    _time.append(re.search("\d{2}:\d{2}", values).group())

#add new columns to the table

data['Date'] = _date

data['Time'] = _time

data.head()
```

```python
#convert ms > minutes
#I should be checking for bad characters, null values, etc.
```

```python
data['Runtime (mins)'] = data['msPlayed'] / 60000
data['Runtime (mins)'] = data['Runtime (mins)'].round(2)

data.head()
```

```python
data['Day'] = pd.to_datetime(data['endTime'])

data['Day'] = data['Day'].dt.day_name()
```

```python
data['endTime'] = pd.to_datetime(data['endTime'])
```
